```python
import numpy as np
import pandas as pd

def count_s(df,st): #func to count strings in a matrix
    return sum([list(df[i]).count(st) for i in df])

def slicing(df,i,j,c,t): 
    return df.loc[i:j,c:t]
def Multiples(value): #to get possible vectors
    return [[i,value/i] for i in range(1,value+1) if value %i==0]
```

```python
F=open("C:\\Users\\murat_gga8ya6\\Desktop\\Desktop\\hash\\pizza\\medium.in","r")
header=F.next()[:-1].split(" ")
data=[]
for lines in F:
    data.append(list(lines[:-1]))
data=pd.DataFrame(data)
data
```

```python
possible_matrix=[]
[possible_matrix.append(Multiples(i)) for i in range(2*int(header[2]),int(header[3])+1)]
#possible_matrix=pd.DataFrame(possible_matrix)
poss=[] # created to obtain a listed version
[[poss.append(j) for j in i] for i in possible_matrix ]
poss.reverse()
poss
```

```python
slices=[]
col_pos=[-1]
startpos=[0,0]
while(1<2):
    for dim in poss:
        t=1
        if (startpos[0]+dim[0]-1<=int(header[0]) and startpos[1],startpos[1]+dim[1]-1<=int(header[1])):
            my_slice=slicing(data,startpos[0],startpos[0]+dim[0]-1,startpos[1],startpos[1]+dim[1]-1)
            if count_s(my_slice,"T") >= int(header[2]) and count_s(my_slice,"M") >= int(header[2]) and ("NaN" not in np.unique(my_slice)):
                t=0
                break
    if t==0:
        data.loc[startpos[0]:startpos[0]+dim[0]-1,startpos[1]:startpos[1]+dim[1]-1]="NaN" #converting pieces to NaN
        slices.append([startpos[0],startpos[0]+dim[0]-1,startpos[1],startpos[1]+dim[1]-1])
    candidates_c=[i for i in range(startpos[1],len(data.columns)) if data.loc[startpos[0],i]!="NaN"]
    #candidates_r=[i for i in range(startpos[0],len(data.index)) if data.loc[i,startpos[1]]!="NaN"]
    
    if len(candidates_c) !=0 and startpos[1]!=int(header[1])-1 : # the bug starts here its about picking the next start point!
        if candidates_c[0]!=startpos[1]:
            startpos[1]=candidates_c[0]
            col_pos.append(startpos[1])
        elif np.max(col_pos)+1 <int(header[1]):
            startpos[1]=np.max(col_pos)+1
            col_pos.append(startpos[1])
    elif startpos[0]!=int(header[0])-1:
        startpos[1]=0
        col_pos=[-1]
        startpos[0]=startpos[0]+1
    else:
        break
```
